```python
import time
start = time.time()

import pymongo
import json
from pymongo import MongoClient
import pprint as pp

client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client.election
mycol = db.electionData
#mycol = db.tnmodi
i=0

for record in mycol.find().limit(20):
    tweet_data={}
    tweet_data['_id']=record['_id']
    i=i+1
    try:
        if('retweeted_status' in record):
            tweet_data['retweet']= True
            print(i)
            if(record['is_quote_status'] == True):
                tweet_data['is_quote_status']= True
                if(record['retweeted_status']['truncated'] == True):
                    tweet_data['quoted_full_text']=record['retweeted_status']['extended_tweet']['full_text']
                    #print('retweeted quoted tweet:\n',record['retweeted_status']['extended_tweet']['full_text'])
                else:
                    tweet_data['quoted_full_text']=record['retweeted_status']['text']
                    #print('retweeted quoted tweet:\n',record['retweeted_status']['text'])
                if(record['retweeted_status']['is_quote_status'] == True):
                    if(record['retweeted_status']['quoted_status']['truncated'] == True):
                        tweet_data['original_full_text']=record['retweeted_status']['quoted_status']['extended_tweet']['full_text']
                        #print('quoted original tweet:\n',record['retweeted_status']['quoted_status']['extended_tweet']['full_text'])
                    else:
                        tweet_data['original_full_text']=record['retweeted_status']['quoted_status']['text']
                        #print('quoted original tweet:\n',record['retweeted_status']['quoted_status']['text'])
            else:
                tweet_data['is_quote_status']= False
                if(record['retweeted_status']['truncated'] == True):
                    tweet_data['original_full_text']=record['retweeted_status']['extended_tweet']['full_text']
                    #print('retweeted original\n',record['retweeted_status']['extended_tweet']['full_text'])
                else:
                    tweet_data['original_full_text']=record['retweeted_status']['text']
                    #print('retweeted original\n',record['retweeted_status']['text'])
            pp.pprint(tweet_data)
        elif(record['is_quote_status'] == True):
            print(i)
            tweet_data['is_quote_status']= True
            tweet_data['retweet'] = False
            if(record['truncated'] == True):
                tweet_data['quoted_full_text'] = record['extended_tweet']['full_text']
                #print('quoted tweet:\n',record['extended_tweet']['full_text'])
            else:
                tweet_data['quoted_full_text'] = record['text'] 
                #print('quoted tweet:\n',record['text'])
            if(record['quoted_status']['truncated'] == True):
                tweet_data['original_full_text'] = record['quoted_status']['extended_tweet']['full_text']
                #print('quoted original tweet:\n',record['quoted_status']['extended_tweet']['full_text'])
            else:
                tweet_data['original_full_text'] = record['quoted_status']['text'] 
                #print('quoted original tweet:\n',record['quoted_status']['text'])
            pp.pprint(tweet_data)
        else:
            tweet_data['retweet'] = False
            tweet_data['is_quote_status'] = False
            print(i) 
            if(record['truncated'] == True):
                tweet_data['original_full_text'] = record['extended_tweet']['full_text'] 
                #print('original \n',record['extended_tweet']['full_text'])
            else:
                tweet_data['original_full_text'] = record['text'] 
                #print('original \n',record['text'])
            pp.pprint(tweet_data)
            
    except:
        continue
        
end = time.time()
print(end-start)
```
